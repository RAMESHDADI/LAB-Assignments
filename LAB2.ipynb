```python
import pandas as pd
a=pd.read_csv('/content/drive/MyDrive/head.csv')
print(a)
```

```python
b=a['Head Size(cm^3)']
min_value=min(b)
max_value=max(b)
print(min_value)
print(max_value)
```

```python
c=[]
for i in range(len(a)):
  d=(b[i]-min_value)/(max_value-min_value)
  c.append(d)
print(c)
```

```python
for i in range(len(a)):
  b[i]=(c[i])
print(b)
```
